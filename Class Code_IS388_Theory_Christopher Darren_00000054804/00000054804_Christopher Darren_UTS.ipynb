```python
data_oth= games[['Other_Sales']]
data_oth
```

```python
#handling outlier with IQR

#Q1, Q3 dan interquartile range
Q1 = data_oth.quantile(q=.25)
Q3 = data_oth.quantile(q=.75)
IQR = Q3-Q1 #IQR=Q3-Q1 or use stats.iqr from scipy

#take value 1.5*IQR from Qi and Q3
data_iqr= data_oth[~((data_oth < (Q1-1.5*IQR)) | (data_oth > (Q3+1.5*IQR))).any(axis=1)]
data_iqr.shape
```

```python
data_zscore=data_oth[(np.abs(stats.zscore(data_oth, nan_policy='omit')) < 3).all(axis=1)]
data_zscore.shape
```

```python
data_iqr.plot(kind='box', figsize=(5,5))
plt.show()
```

```python
data_zscore.columns = ['Z-Score']
data_iqr.columns=['IQR']
```

```python
result=pd.concat([data_oth, data_zscore,data_iqr], axis=1)
```

```python
fig, ax= plt.subplots(figsize=(10,10))
sns.boxplot(data=result)
```
